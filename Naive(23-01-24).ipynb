```python
import numpy as np
import matplotlib.pyplot as plt

# Toy dataset: two features (X1, X2), two classes (0, 1)
data = np.array([
    [2, 3, 0],
    [1, 2, 0],
    [3, 4, 0],
    [6, 5, 1],
    [7, 7, 1],
    [5, 6, 1]
])

# Separate data by class
class_0 = data[data[:, 2] == 0]
class_1 = data[data[:, 2] == 1]

# Calculate class means and standard deviations
mean_0 = np.mean(class_0[:, :2], axis=0)
std_0 = np.std(class_0[:, :2], axis=0)

mean_1 = np.mean(class_1[:, :2], axis=0)
std_1 = np.std(class_1[:, :2], axis=0)

# Function to calculate probability density function of a normal distribution
def gaussian_pdf(x, mean, std):
    return (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std)**2)

# Plot the data points
plt.scatter(class_0[:, 0], class_0[:, 1], label='Class 0', marker='o')
plt.scatter(class_1[:, 0], class_1[:, 1], label='Class 1', marker='x')

# Plot decision boundary
x_vals = np.linspace(0, 8, 100)
y_vals = np.linspace(2, 8, 100)

for x in x_vals:
    for y in y_vals:
        prob_class_0 = np.prod(gaussian_pdf(np.array([x, y]), mean_0, std_0))
        prob_class_1 = np.prod(gaussian_pdf(np.array([x, y]), mean_1, std_1))

        if prob_class_0 > prob_class_1:
            plt.scatter(x, y, color='blue', alpha=0.1)
        else:
            plt.scatter(x, y, color='orange', alpha=0.1)

plt.title('Naive Bayes Classifier Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
